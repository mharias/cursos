```python
def lg(mu,n, ybar):
    mu2 = mu**2
    log_g_mu = n * (y_bar*mu-mu2/2.0)-np.log(1+mu2)
    return log_g_mu
```

```python
def mh(n,ybar,n_iter,mu_init, candidate_sd):
    mu_out = np.zeros(n_iter)
    accept = 0
    mu_now = mu_init
    lg_now = lg(mu=mu_now,n=n,ybar=ybar)
    for i in range(n_iter):
        mu_candidate = norm.rvs(loc = mu_now, scale = candidate_sd, size=1)
        lg_candidate = lg(mu=mu_candidate, n=n, ybar=ybar)
        alpha = np.exp(lg_candidate-lg_now) #al ser logaritmos calculamos la exponencial
        u = uniform.rvs()
        if u < alpha:
            mu_now = mu_candidate
            accept+=1
            lg_now = lg_candidate
        mu_out[i] = mu_now
    return mu_out,accept/n_iter
```
