```python
def lg(mu,n, ybar):
    mu2 = mu**2
    log_g_mu = n * (y_bar*mu-mu2/2.0)-np.log(1+mu2)
    return log_g_mu
```

```python
def mh(n,ybar,n_iter,mu_init, candidate_sd):
    mu_out = np.zeros(n_iter)
    accept = 0
    mu_now = mu_init
    lg_now = lg(mu=mu_now,n=n,ybar=ybar)
    for i in range(n_iter):
        mu_candidate = norm.rvs(loc = mu_now, scale = candidate_sd, size=1)
        lg_candidate = lg(mu=mu_candidate, n=n, ybar=ybar)
        alpha = np.exp(lg_candidate-lg_now) #al ser logaritmos calculamos la exponencial
        u = uniform.rvs()
        if u < alpha:
            mu_now = mu_candidate
            accept+=1
            lg_now = lg_candidate
        mu_out[i] = mu_now
    return mu_out,accept/n_iter
```

```python
y = np.array([1.2,1.4,-.5,.3,.9,2.3,1.0,0.1,1.3,1.9])
```

```python
y_bar = y.mean()
```

### 3 posteriors

## Número 0

```python
n_iter=10000
```

```python
post0, exito = mh(n=n,ybar=y_bar,n_iter=n_iter,mu_init=0, candidate_sd=0.9)
```

```python
print ('Tasa aceptacion {}'.format(exito))
```

```python
plt.plot(range(n_iter),post0)
```

```python
corr0 = plot_acf(post0, lags = 40)
```

```python
pm.effective_n(post0)
```
